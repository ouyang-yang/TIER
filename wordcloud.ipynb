```python
import jieba
import jieba.analyse
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from PIL import Image
from collections import Counter
from ArticutAPI import Articut
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
txt110 = "database/110_raw_policyInfo.csv"
text110 = open(txt110, 'r', encoding='utf-8-sig').read()
txt111 = "database/111_raw_policyInfo.csv"
text111 = open(txt111, 'r', encoding='utf-8-sig').read()

dictfile = "C:/Users/mayda/AppData/Local/Programs/Python/Python39/Lib/site-packages/jieba/dict.txt"
fontpath = "C:/Users/mayda/Downloads/genseki-font-master/genseki-font-master/ttc/GenSekiGothic-R.ttc"
stopwords = "C:/Users/mayda/AppData/Local/Programs/Python/Python39/Lib/site-packages/jieba/my_dict.txt"
### 繁體字較完整詞庫
### https://raw.githubusercontent.com/ldkrsi/jieba-zh_TW/master/jieba/dict.txt
```

```python
jieba.set_dictionary(dictfile)

tags = jieba.analyse.extract_tags(text110, topK=25)
seg_list = jieba.lcut(text110, cut_all=False)
dictionary = Counter(seg_list)

freq = {}
for word in dictionary:
    if word in tags and word not in stopwords:
        freq[word] = dictionary[word]
print(freq)
```

```python
wordcloud = WordCloud(
    background_color="white", contour_color="steelblue", font_path=fontpath).generate_from_frequencies(freq)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
